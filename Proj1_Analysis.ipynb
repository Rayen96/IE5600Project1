```python
# Libraries for data analysis & visualization

import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
import seaborn as sns

#stock data https://www.theancientbabylonians.com/using-python-to-analyse-singapore-banking-stocks-dbs-uob-ocbc/

tickers = ['T39.SI', 'S58.SI', 'U96.SI'] #Singapore Press Holdings Limited, SATS Ltd., Sembcorp Industries Ltd
mydata = pd.DataFrame()
for ticker in tickers:
    mydata[ticker] = wb.DataReader(ticker, data_source='yahoo', start='2010-1-1')['Adj Close']
```

```python
mydata
```

```python
mydata.plot(figsize=(10,6))
plt.legend(["Singapore Press Holdings Limited", "SATS Ltd.", "Sembcorp Industries Ltd"])
plt.xlabel("Date")
plt.ylabel("Price in SGD")
plt.show()
```

```python
#normalize data due to different starting prices from 2005

(mydata/mydata.iloc[0]*100).plot(figsize=(10,6));

plt.legend(["Singapore Press Holdings Limited", "SATS Ltd.", "Sembcorp Industries Ltd"])
plt.xlabel("Date")
plt.ylabel("Normalized Daily Adjusted Closing Price (%)")
plt.grid()
plt.show()
```

```python
#measuring annual log returns in %

mydata_log = np.log(mydata/mydata.shift(1)) #current price divided by previous day's price & get log return
mydata_log
```

```python
round((mydata_log[['T39.SI', 'S58.SI', 'U96.SI']].mean() * 252 * 100), 3)
#SATS Ltd has the highest annual log-returns followed by Sembcorp & lastly Singapore Press Holdings Limited.
```

```python
#measure annual log volatility in % by comparing the annualised standard deviation.
#sd refers to how far do prices deviate away from the mean. The higher the sd, the more volatile a stock is.

#annual log volatility in %

round((mydata_log[['T39.SI', 'S58.SI', 'U96.SI']].std() * (252**0.5) * 100), 3)

#highest volatility: Sembcorp followed by SATS Ltd. Least volatility: Singapore Press Holdings Limited. 
```

```python
#volatility chart

vol = mydata_log.rolling(200).std() * np.sqrt(200) #200-day rolling average, standard deviation line
vol.plot(figsize=(10,6))

plt.legend(["Singapore Press Holdings Limited", "SATS Ltd.", "Sembcorp Industries Ltd"])
plt.xlabel("Date")
plt.ylabel("SD")
plt.show()

#least volatile is Singapore Press Holdings compared to SATS & Sembcorp.
```

```python
Annual_log_returns = round((mydata_log[['T39.SI', 'S58.SI', 'U96.SI']].mean() * 252 * 100), 3)
Annual_log_volatility = round((mydata_log[['T39.SI', 'S58.SI', 'U96.SI']].std() * (252**0.5) * 100), 3)

summary = {
    'SG Blue-Chip Companies': ['Singapore Press Holdings Limited (SGX: T39)', 'Sembcorp Industries Limited (SGX: U96)', 'SATS Ltd (SGX: S58)'],
    'Annual % Returns': (Annual_log_returns),
    'Annual % Volatility': (Annual_log_volatility)
        
}

pd.DataFrame(summary).set_index('SG Blue-Chip Companies')
```
